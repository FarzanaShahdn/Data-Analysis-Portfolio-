```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv('netflix_titles.csv')

#Display the first 5 rows
df.head()
```

```python
#Shape and info
print ("Shape:", df.shape)
df.info()

#Check missing value
df.isnull().sum()

#Check duplicate
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
#Drop rows with missing values in 'title'
df+df.dropna(subset=['title'])
```

```python
import pandas as pd

type_counts = df['type'].value_counts().reset_index()
type_counts.columns = ['Type', 'Count']

# Display the table
type_counts
```

```python
#How many shows are Movies vs TV Shows?

df["type"].value_counts().plot(kind='bar', title ='Movies vs Tv Shows')
plt.xlabel("Type")
plt.ylabel ("Count")
plt.show()
```

### **Summary: Movies vs TV Shows**

- The dataset reveals that **Movies** significantly outnumber **TV Shows** on Netflix.
- There are over **6,000 Movies**, while **TV Shows** total around **2,700–3,000**.
- This suggests that **Netflix's content strategy leans more toward movies than serialized shows.**

```python
#Top 10 countries producing the most content

top_countries =df[' country'].value_counts().head(10)
top_countries.plot(kind='bar', title= "Top 10 Countries Producing the Most Content")
plt.ylabel('Number of Titles')
plt.show()
```

**Summary: Top 10 Countries Producing the Most Netflix Content**

- The **United States** dominates content production on Netflix with nearly **3,000 titles**.

- **India** is the second-largest contributor with close to **1,000 titles**.

- Other top contributors include the **United Kingdom, Japan, and South Korea.**

- Countries like **Canada, Spain, France, Mexico, and Egypt** also make the top 10 list.

- This distribution **highlights Netflix's strong focus on North America and Asia,** while steadily expanding its international content catalog.

```python
#How many Adult (18+) shows?

#Check ratings
df['rating'].value_counts()

#Filter for 18+ content
adult = df[df['rating'].isin(['TV-MA','R','NC-17'])]
print("Total 18+ Content",adult.shape[0])
```

### Summary : How many Adult (18+) shows?

- *There are about **4009** adult (18+) rated shows on Netflix.*

```python
import pandas as pd

# Filter for adult (18+) ratings
adult_ratings = ['TV-MA', 'R', 'NC-17']  # Common 18+ ratings
adult_movies = df[(df['type'] == 'Movie') & (df['rating'].isin(adult_ratings))]

# Remove rows with missing country
adult_movies = adult_movies.dropna(subset=['country'])

# Handle multiple countries (some rows have comma-separated values)
# Split and explode to count each country separately
adult_movies['country'] = adult_movies['country'].str.split(', ')
adult_movies = adult_movies.explode('country')

# Count top 10 countries producing 18+ movies
top_adult_countries = adult_movies['country'].value_counts().head(10)

# Convert to DataFrame for display
top_adult_countries_df = top_adult_countries.reset_index()
top_adult_countries_df.columns = ['Country', 'Number of 18+ Movies']

# Display table
top_adult_countries_df
```

```python
import matplotlib.pyplot as plt

top_adult_countries.plot(kind='bar', figsize=(10,5), title='Top 10 Countries Producing 18+ Movies')
plt.xlabel("Country")
plt.ylabel("Number of 18+ Movies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### **Summary: Top 10 Countries Producing 18+ Rated Movies on Netflix**

- The **United States** dominates 18+ movie production on Netflix, with nearly **1,400 titles**, far surpassing other countries.

- The **United Kingdom** and **India** follow at a distant second and third, with around **280** and **240** titles respectively.

- European countries like **France, Germany, Spain, and Belgium** also contribute notably to mature-rated content.

- **Canada, Mexico, and Argentina** round out the top 10, reflecting Netflix inclusion of adult content from across North and South America.

- This distribution suggests that adult-rated content on Netflix is largely driven by **Western and English-speaking markets.**

```python
#Number of releases over the years

df['release_year'].value_counts().sort_index().plot(kind= 'line', figsize = (10,5))
plt.title("Netfilx of releases over the years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid (True)
plt.show()
```

### **Summary: Netflix Releases Over the Years**

- Netflix content releases remained minimal before the year **2000**, with very few titles added each year.

- A noticeable increase began in the early **2010s**, accelerating rapidly around **2015**.

- The peak year appears to be around **2019–2020**, with Netflix releasing over **1,000 titles** in a single year.

- A slight decline is visible after the peak, possibly due to the **COVID-19 pandemic’s impact** on production and release schedules.

- Overall, the trend reflects Netflix’s aggressive content expansion in recent years, especially during the **streaming boom era.**
